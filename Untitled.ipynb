```python
import re
import pandas as pd
```

```python
f=open('yogi.txt','r',encoding='utf-8')
```

```python
data=f.read()
```

```python
pattern=r'\d{1,2}\/\d{1,2}\/\d{2},\s\d{1,2}:\d{2}\s(?:AM|PM)\s-\s'
```

```python
messages=re.split(pattern,data)[1:]
messages
```

```python
date=re.findall(pattern,data)
date
```

```python
df=pd.DataFrame({'user_message':messages ,'message_date':date})
df['message_date']=df['message_date'].str.replace(r'\u202f', ' ')
```

```python
df.head()
```

```python
def add_2000_to_year(row):
    date_part = row.split(',')[0]  
    date_parts = date_part.split('/') 
    
    year = int(date_parts[2])  
    if year < 100:  
        date_parts[2] = str(2000 + year)  
    
    full_date = '/'.join(date_parts)
    
    row = row.replace(date_part, full_date)
    
    return row

df['message_date'] = df['message_date'].apply(add_2000_to_year)
```

```python
from datetime import datetime
```

```python
times=df.iloc[:,-1]
```

```python
times.head()
```

```python
converted_timestamps = times.apply(
    lambda row: datetime.strptime(row.rstrip(" -"), "%m/%d/%Y, %I:%M %p").strftime("%d/%m/%Y %H:%M")
)

# Display the converted timestamps
for ts in times:
    print(ts)
```

```python
df = pd.DataFrame({
    "Date": converted_timestamps,
    "Message": df.iloc[:,0]
})
```

```python
df['Date']
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y %H:%M', errors='coerce')
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y %H:%M')
```

```python
import pandas as pd
import re

# Example DataFrame

# Lists to store extracted users and messages
users = []
messages = []

# Loop through the messages
for message in df['Message']:
    entry = re.split(r'([\w\W]+?):\s', message)  # Corrected regex pattern
    if len(entry) > 1:  # There is a user in the message
        users.append(entry[1])  # Extract username
        messages.append(entry[2])  # Extract message
    else:  # For messages without a username (e.g., group notifications)
        users.append('group_notification')  # Assign group notification
        messages.append(entry[0])  # The entire message

# Add the new columns to the DataFrame
df['user'] = users
df['message'] = messages

# Drop the original 'Message' column
df.drop(columns=['Message'], inplace=True)

# Display the updated DataFrame
print(df)
```
